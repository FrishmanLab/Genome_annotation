```python
# FDR =  decoys / trargets
df = df.sort_values('Score', ascending=False)
len_decoys = 0
len_targets = 0
FDR_dict = {}
for index, row in df.iterrows():
    if row['Leading razor protein'].startswith('REV_'):
        len_decoys += 1
    else:
        len_targets+=1
    
    FDR = len_decoys/len_targets
    Score = row['Score']

    FDR_dict[Score] = FDR
print(f'decoys = {len_decoys}')
print(f'targets = {len_targets}')
print(FDR_dict)


FDR_dict_filterd = {}
for key, value in FDR_dict.items():
    if value >= 0.01 and value <= 0.011:
        FDR_dict_filterd[key] = value
print(FDR_dict_filterd)
print(min(FDR_dict_filterd.values()))
for k, v in FDR_dict_filterd.items():
    if v == min(FDR_dict_filterd.values()):
        print(k)
```
